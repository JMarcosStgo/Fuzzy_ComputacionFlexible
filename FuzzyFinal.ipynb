```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

Crear las variables de entrada y salida

```python
calidad_comida = ctrl.Antecedent(np.arange(0, 10, 0.1), 'calidad_comida')
tiempo_espera = ctrl.Antecedent(np.arange(0, 30, 0.1), 'tiempo_espera')
amabilidad_personal = ctrl.Antecedent(np.arange(0, 10, 0.1), 'amabilidad_personal')
satisfaccion_cliente = ctrl.Consequent(np.arange(0, 100, 0.1), 'satisfaccion_cliente')
```

Definicion de las funciones de membresia

```python
calidad_comida['baja'] = fuzz.trapmf(calidad_comida.universe, [0, 0, 3, 4])
calidad_comida['media'] = fuzz.trapmf(calidad_comida.universe, [2, 4, 6, 8])
calidad_comida['alta'] = fuzz.trapmf(calidad_comida.universe, [6, 7, 10, 10])
calidad_comida.view()
```

```python
tiempo_espera['corto'] = fuzz.trapmf(tiempo_espera.universe, [0, 0, 7, 10])
tiempo_espera['medio'] = fuzz.trapmf(tiempo_espera.universe, [5, 10, 20, 25])
tiempo_espera['largo'] = fuzz.trapmf(tiempo_espera.universe, [20, 23, 30, 30])
tiempo_espera.view()
```

```python
amabilidad_personal['baja'] = fuzz.trapmf(amabilidad_personal.universe, [0, 0, 3, 4])
amabilidad_personal['media'] = fuzz.trapmf(amabilidad_personal.universe, [2, 4, 6, 8])
amabilidad_personal['alta'] = fuzz.trapmf(amabilidad_personal.universe, [6, 7, 10, 10])
amabilidad_personal.view()
```

```python
satisfaccion_cliente['baja'] = fuzz.trapmf(satisfaccion_cliente.universe, [0, 0, 20, 30])
satisfaccion_cliente['media'] = fuzz.trapmf(satisfaccion_cliente.universe, [20, 30, 70, 80])
satisfaccion_cliente['alta'] = fuzz.trapmf(satisfaccion_cliente.universe, [70, 80, 100, 100])
satisfaccion_cliente.view()
```

Reglas difusas

```python
rule1 = ctrl.Rule(calidad_comida['baja'] & tiempo_espera['corto'] & amabilidad_personal['baja'], satisfaccion_cliente['baja'])
rule2 = ctrl.Rule(calidad_comida['baja'] & tiempo_espera['corto'] & amabilidad_personal['media'], satisfaccion_cliente['baja'])
rule3 = ctrl.Rule(calidad_comida['baja'] & tiempo_espera['corto'] & amabilidad_personal['alta'], satisfaccion_cliente['baja'])
rule4 = ctrl.Rule(calidad_comida['baja'] & tiempo_espera['medio'] & amabilidad_personal['baja'], satisfaccion_cliente['baja'])
rule5 = ctrl.Rule(calidad_comida['baja'] & tiempo_espera['medio'] & amabilidad_personal['media'], satisfaccion_cliente['baja'])
rule6 = ctrl.Rule(calidad_comida['baja'] & tiempo_espera['medio'] & amabilidad_personal['alta'], satisfaccion_cliente['baja'])
rule7 = ctrl.Rule(calidad_comida['baja'] & tiempo_espera['largo'] & amabilidad_personal['baja'], satisfaccion_cliente['baja'])
rule8 = ctrl.Rule(calidad_comida['baja'] & tiempo_espera['largo'] & amabilidad_personal['media'], satisfaccion_cliente['baja'])
rule9 = ctrl.Rule(calidad_comida['baja'] & tiempo_espera['largo'] & amabilidad_personal['alta'], satisfaccion_cliente['baja'])
rule10 = ctrl.Rule(calidad_comida['media'] & tiempo_espera['corto'] & amabilidad_personal['baja'], satisfaccion_cliente['baja'])
rule11 = ctrl.Rule(calidad_comida['media'] & tiempo_espera['corto'] & amabilidad_personal['media'], satisfaccion_cliente['media'])
rule12 = ctrl.Rule(calidad_comida['media'] & tiempo_espera['corto'] & amabilidad_personal['alta'], satisfaccion_cliente['media'])
rule13 = ctrl.Rule(calidad_comida['media'] & tiempo_espera['medio'] & amabilidad_personal['baja'], satisfaccion_cliente['media'])
rule14 = ctrl.Rule(calidad_comida['media'] & tiempo_espera['medio'] & amabilidad_personal['media'], satisfaccion_cliente['media'])
rule15 = ctrl.Rule(calidad_comida['media'] & tiempo_espera['medio'] & amabilidad_personal['alta'], satisfaccion_cliente['media'])
rule16 = ctrl.Rule(calidad_comida['media'] & tiempo_espera['largo'] & amabilidad_personal['baja'], satisfaccion_cliente['media'])
rule17 = ctrl.Rule(calidad_comida['media'] & tiempo_espera['largo'] & amabilidad_personal['media'], satisfaccion_cliente['media'])
rule18 = ctrl.Rule(calidad_comida['media'] & tiempo_espera['largo'] & amabilidad_personal['alta'], satisfaccion_cliente['media'])
rule19 = ctrl.Rule(calidad_comida['alta'] & tiempo_espera['corto'] & amabilidad_personal['baja'], satisfaccion_cliente['media'])
rule20 = ctrl.Rule(calidad_comida['alta'] & tiempo_espera['corto'] & amabilidad_personal['media'], satisfaccion_cliente['media'])
rule21 = ctrl.Rule(calidad_comida['alta'] & tiempo_espera['corto'] & amabilidad_personal['alta'], satisfaccion_cliente['alta'])
rule22 = ctrl.Rule(calidad_comida['alta'] & tiempo_espera['medio'] & amabilidad_personal['baja'], satisfaccion_cliente['media'])
rule23 = ctrl.Rule(calidad_comida['alta'] & tiempo_espera['medio'] & amabilidad_personal['media'], satisfaccion_cliente['alta'])
rule24 = ctrl.Rule(calidad_comida['alta'] & tiempo_espera['medio'] & amabilidad_personal['alta'], satisfaccion_cliente['alta'])
rule25 = ctrl.Rule(calidad_comida['alta'] & tiempo_espera['largo'] & amabilidad_personal['baja'], satisfaccion_cliente['alta'])
rule26 = ctrl.Rule(calidad_comida['alta'] & tiempo_espera['largo'] & amabilidad_personal['media'], satisfaccion_cliente['alta'])
rule27 = ctrl.Rule(calidad_comida['alta'] & tiempo_espera['largo'] & amabilidad_personal['alta'], satisfaccion_cliente['alta'])
```

Sistema difuso

```python
satisfaction_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20, rule21, rule22, rule23, rule24, rule25, rule26, rule27])
satisfaction = ctrl.ControlSystemSimulation(satisfaction_ctrl)
```

Entradas para la regla1

```python
satisfaction.input['calidad_comida'] = 1
satisfaction.input['tiempo_espera'] = 8
satisfaction.input['amabilidad_personal'] = 3
```

Computar el resultado y mostrar el resultado

```python
satisfaction.compute()
print("Nivel de Satisfaccin:", satisfaction.output['satisfaccion_cliente'])
satisfaccion_cliente.view(sim=satisfaction)
```
